## Hangman

```python
import random 

words = []

#Read the list of words
with open('sowpods.txt', 'r') as f:
    line = f.readline().strip()
    words.append(line)
    while line: 
        line = f.readline().strip()
        words.append(line)

#Generate a random number
rand_index = random.randint(0, len(words))

#Print the word
#print('Random Word: ', words[rand_index])  

```

```python
print('Welcome to Hangman!')

word = words[rand_index]
clue_word = ['_'] * len(word)
playing = True
guessed_letters = set()
nr_guesses = 6

while playing: 

    print(clue_word)
    
    if '_' not in clue_word:
        print('\nYou won! Congratulations!')
        break
        
    guess = str(input('\n>>> Guess your letter: ')).upper()
    
    if guess in guessed_letters:
        print('\nLetter already guessed!')
        #nr_guesses+=1
        continue  
    else:
        for i in range(len(word)):
            if guess == word[i]:
                clue_word[i] = guess
        nr_guesses-=1
    
    if guess in clue_word:    
        nr_guesses+=1
        
    guessed_letters.add(guess)
    
    print('You have ' + str(nr_guesses) + ' guesses left')
    
    if nr_guesses == 0:
        print('Game over. The word was ' + str(word))
        break
```
